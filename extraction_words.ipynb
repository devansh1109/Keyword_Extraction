```python
def remove_punctuations(text):
    pattern = r'[!"#$%&\'()*+,\-./:;<=>?@[\\]^_`{|}~]'
    return re.sub(pattern, '', text)

# Apply the function to the specified column, handling NaN values
df_selected1['abstract'] = df_selected1['abstract'].apply(lambda x: remove_punctuations(str(x)) if pd.notnull(x) else x)
```

```python
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    if isinstance(text, str):
        words = text.split()
        filtered_words = [word for word in words if word.lower() not in stop_words]
        return ' '.join(filtered_words)
    else:
        return text
    
# Apply the stop word removal function to the specified column
df_selected1['abstract'] = df_selected1['abstract'].apply(remove_stopwords)

# # Replace empty strings with NaN
df_selected1['abstract'].replace('', np.nan, inplace=True)

# Save the modified DataFrame back to a CSV file
df_selected1.to_csv('df_selected', index=False)

df_selected1.head()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Specify the column name containing the abstracts
column_name = 'abstract'

# Drop rows with NaN values in 'abstract' column
df_selected1 = df_selected1.dropna(subset=[column_name]).reset_index(drop=True)

# Initialize TfidfVectorizer with customized parameters
vectorizer = TfidfVectorizer(
    stop_words='english',       # Remove common English stop words
    token_pattern=r'\b[a-zA-Z]{4,}\b',  # Match words of 3 or more alphanumeric characters
    ngram_range=(1, 3),         # Include unigrams and bigrams
    max_df=0.1,    
                
)

# Fit and transform the abstracts into TF-IDF matrix
tfidf_matrix = vectorizer.fit_transform(df_selected1[column_name])

# Get the feature names (words)
feature_names = vectorizer.get_feature_names_out()

# Function to extract top N keywords for each abstract
def extract_top_keywords(row, top_n=5):
    try:
        row_index = row.name  # Get the row index
        tfidf_scores = tfidf_matrix[row_index].toarray()[0]
        tfidf_df = pd.DataFrame({'word': feature_names, 'tfidf': tfidf_scores})
        tfidf_df = tfidf_df.sort_values(by='tfidf', ascending=False)
        top_keywords = tfidf_df.head(top_n)['word'].values
        return ', '.join(top_keywords)
    except IndexError:
        return None  # Handle index out of range due to NaN values

# Extract top keywords for each abstract and store in a new column 'keywords'
df_selected1['keywords'] = df_selected1.apply(extract_top_keywords, axis=1)

# Save the DataFrame with the new 'keywords' column to a CSV file
df_selected1.to_csv('df_selected_with_keywords2.csv', index=False)

# Display the DataFrame to verify the results
print(df_selected1[['abstract', 'keywords']].head())
```
